```python
from getpass import getpass
selecao = 1
entrada = 0

def menu_selecao():
  print("Bem vindo ao Unifood")
  print("- Veja as opções -")
  print("0-Fechar o aplicativo")
  print("1-Realize o cadastro do usuário")
  print("2-Realizar login")
  selecao = input("Selecione o número de uma das opções acima para continuar: ")
  return (selecao)

def cadastro():
  info_user = open('info_usua.txt', 'r')
  lista_info = info_user.readlines()
  info_user.close()
  while(True):
    email = input("Digite o seu e-mail: ")
    login = 0
    for i in lista_info:
      info = i.strip().split(';')
      if info[0] == email:
        print('Este e-mail já está em uso, tente usar outro')
        login = 1
    if login != 1:
      info = open('info_usua.txt', 'a')
      info.write(email)
      info.close()
      break
  with open('info_usua.txt', 'a') as info:
    info.write(";")
    info.write(getpass("Digite sua senha: "))
    info.write("\n")
  info.close()
  print("Cadastro feito com sucesso")

def cadastro_inicial():
  with open('info_usua.txt', 'a') as info:
    info.write(input("Digite o seu e-mail: "))
    info.write(";")
    info.write(getpass("Digite sua senha: "))
    info.write("\n")
  info.close()
  print("Cadastro feito com sucesso")

def login_usuario():
  login = 0
  while True:
    LoginEmail = input("Digite o seu e-mail: ")
    LoginSenha = getpass("Digite sua senha: ")
    info_user = open('info_usua.txt', 'r')
    lista_info = info_user.readlines()
    info_user.close()
    for i in lista_info:
      info = i.strip().split(';')
      if LoginEmail == info[0] and LoginSenha == info[1]:
        login = 1
    if login == 1:
      print("Login efetuado com sucesso")
      break
    else:
      print("E-mail ou senha incorreta, tente novamente")

while(True):
  opcao = menu_selecao()
  if opcao == '0':
    print('Fechando aplicativo...')
    break
  elif entrada == 1 and opcao == '1':
    resultado = cadastro()
  elif opcao == '1':
    resultado = cadastro_inicial()
    entrada = 1
  elif opcao == '2':
    resultado = login_usuario()
```
